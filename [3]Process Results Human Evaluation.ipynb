```python
import pandas as pd
```

```python
df = pd.read_excel('Final_Results_Survey.xlsx',
                  usecols=(list(range(1,115))))
```

```python
df = df.iloc[1:]
```

```python
df_list = []
```

```python
df_list.append(df[['Duration (in seconds)'] + 
                [c[i] for i,item in enumerate(c) if "Q4" in item] + 
                ['Q5.2', 'Q5.2_3_TEXT', 'Q5.3', 'Q5.4', 'Q113', 'Q113_6_TEXT', 'Q5.5', 'Q5.6']])
```
